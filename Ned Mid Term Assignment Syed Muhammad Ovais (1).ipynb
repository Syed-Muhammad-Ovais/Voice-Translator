```python
import tkinter as tk
from tkinter import ttk, messagebox
import speech_recognition as sr
from googletrans import Translator, LANGUAGES
from gtts import gTTS
import os
import tempfile
import threading
import time


def voice_to_text():
    recognizer = sr.Recognizer()

    with sr.Microphone() as source:
        recognizer.adjust_for_ambient_noise(source)  # adjust for background noises
        audio = recognizer.listen(source)
        try:
            text = recognizer.recognize_google(audio)
            return text
        except sr.UnknownValueError:
            messagebox.showerror("Error", "Sorry, I did not hear your request.")
        except sr.RequestError:
            messagebox.showerror("Error", "Sorry, the API is unavailable.")
    return None

def text_to_voice(text, lang='en'):
    tts = gTTS(text=text, lang=lang)
    with tempfile.NamedTemporaryFile(delete=True) as fp:
        tts.save(fp.name + ".mp3")
        os.system(f"start {fp.name}.mp3")
        time.sleep(5)

def translate_text(text, target_language='en'):
    translator = Translator()
    translation = translator.translate(text, dest=target_language)
    return translation.text

def perform_translation():
    source_text = source_entry.get()
    target_language = target_language_combobox.get()

    if not source_text:
        messagebox.showwarning("Warning", "Please enter or say something to translate!")
        return

    translated_text = translate_text(source_text, target_language)
    target_text.set(translated_text)

    threading.Thread(target=text_to_voice, args=(translated_text, target_language)).start()

def start_listening():
    source_text = voice_to_text()
    source_entry.delete(0, tk.END)
    source_entry.insert(0, source_text)
    perform_translation()


app = tk.Tk()
app.title("Translator App")

frame = ttk.Frame(app, padding="10")
frame.grid(row=0, column=0, sticky=(tk.W, tk.E, tk.N, tk.S))

source_text = tk.StringVar()
source_label = ttk.Label(frame, text="Speak or type text:")
source_label.grid(column=0, row=0, sticky=tk.W, padx=5, pady=5)

source_entry = ttk.Entry(frame, width=40, textvariable=source_text)
source_entry.grid(column=1, row=0, padx=5, pady=5)

listen_button = ttk.Button(frame, text="Listen", command=start_listening)
listen_button.grid(column=2, row=0, padx=5, pady=5)

target_text = tk.StringVar()
target_label = ttk.Label(frame, text="Translated text:")
target_label.grid(column=0, row=1, sticky=tk.W, padx=5, pady=5)

target_entry = ttk.Entry(frame, width=40, textvariable=target_text, state='readonly')
target_entry.grid(column=1, row=1, padx=5, pady=5)

language_label = ttk.Label(frame, text="Target Language:")
language_label.grid(column=0, row=2, sticky=tk.W, padx=5, pady=5)

target_language_combobox = ttk.Combobox(frame, values=list(LANGUAGES.values()))
target_language_combobox.grid(column=1, row=2, padx=5, pady=5)
target_language_combobox.set('english')

translate_button = ttk.Button(frame, text="Translate", command=perform_translation)
translate_button.grid(column=2, row=2, padx=5, pady=5)

app.mainloop()
```
